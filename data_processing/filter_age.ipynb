## 分區

```python
fc_year = np.floor(fc['age (month)']/12)
kd_year = np.floor(kd['age (month)']/12)
fc_set = set(fc_year)
kd_set = set(kd_year)
fc_year_count = {}
kd_year_count = {}
for age in fc_set:
    fc_year_count[age] = list(fc_year).count(age)

for age in fc_set:
    kd_year_count[age] = list(kd_year).count(age)
```

```python
fc_year_count_df_region = pd.DataFrame.from_dict(fc_year_count,orient='index')
fc_year_count_df_region = fc_year_count_df_region.T
fc_year_count_df_region.loc[1] = fc_year_count_df_region.loc[0]/56454
#fc_year_count_df_region.to_csv('D:/CODE/KD/code_after_web/data/KeptGPT/fc_year_count_region.csv',index=False)
kd_year_count_df_region = pd.DataFrame.from_dict(kd_year_count,orient='index')
kd_year_count_df_region = kd_year_count_df_region.T
kd_year_count_df_region.loc[1] = kd_year_count_df_region.loc[0]/1815
#kd_year_count_df_region.to_csv('D:/CODE/KD/code_after_web/data/KeptGPT/kd_year_count_region.csv',index=False)
```
